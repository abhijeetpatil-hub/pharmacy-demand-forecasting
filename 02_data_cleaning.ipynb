```python
import pandas as pd
```

```python
df = pd.read_csv("../data/pharmacy_sales.csv")
df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Expiry_Date"] = pd.to_datetime(df["Expiry_Date"])
```

```python
df.isnull().sum()
```

```python
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Expiry_Days"] = (df["Expiry_Date"] - df["Date"]).dt.days

df.head()
```

```python
df.to_csv("../data/pharmacy_sales_cleaned.csv", index=False)
```
