```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
dates = pd.date_range(start="2023-01-01", end="2024-12-31", freq="D")
```

```python
medicines = ["Paracetamol", "Azithromycin", "Amoxicillin",
             "Cetrizine", "Pantoprazole", "Metformin", "Atorvastatin"]

categories = ["Tablet", "Capsule", "Syrup"]
cities = ["Mumbai", "Pune", "Delhi", "Bangalore", "Hyderabad"]
suppliers = ["Cipla", "Sun Pharma", "Dr Reddy", "Abbott"]
```

```python
data = []

for date in dates:
    med = np.random.choice(medicines)
    units = np.random.randint(5, 200)
    price = np.random.randint(5, 50)
    stock = np.random.randint(100, 1000)

    data.append([
        date,
        med,
        np.random.choice(categories),
        units,
        price,
        units * price,
        stock,
        np.random.choice(suppliers),
        date + pd.Timedelta(days=np.random.randint(180, 720)),
        np.random.choice(cities)
    ])
```

```python
columns = [
    "Date", "Medicine_Name", "Category", "Units_Sold",
    "Unit_Price", "Revenue", "Stock_Available",
    "Supplier", "Expiry_Date", "City"
]

df = pd.DataFrame(data, columns=columns)
```

```python
df.to_csv("../data/pharmacy_sales.csv", index=False)
df.head()
```
